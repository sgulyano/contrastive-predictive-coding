```python
from maskdata_utils import FaceMaskTileGenerator, FaceMaskGenerator

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
batch_size = 16
train_data = FaceMaskTileGenerator(batch_size=batch_size, subset='train',
                                   positive_samples=batch_size // 2, rescale=True)
```

```python
x, y = next(train_data)

plt.figure(figsize=[16,8])
for i in range(4):
    plt.subplot(2,4,i+1)
    plt.imshow((x[0][0][i] + 1.)/2.)
for i in range(3):
    plt.subplot(2,4,i+5)
    plt.imshow((x[1][0][i] + 1.)/2.)
    plt.title(y[0])
plt.show()
```
